```python
from pathlib import Path


def move_and_symlink(file_location, directories):
    file_path = Path(file_location)
    new_location = Path(directories[0]) / file_path.name
    try:
        # Move the file to the new location
        new_location.parent.mkdir(parents=True, exist_ok=True)
        file_path.rename(new_location)
    except Exception as e:
        print(f"Failed to move {file_path} to {new_location}: {e}")
        return
    # Create symlinks in the other directories
    for directory in directories[1:]:
        symlink_path = Path(directory) / file_path.name
        try:
            # symlink_path.parent.mkdir(parents=True, exist_ok=True)
            symlink_path.symlink_to(new_location)
        except Exception as e:
            print(f"Failed to create symlink at {symlink_path}: {e}")


file_location = "path/to/your/file.txt"
directories = ["path/to/new/location", "path/to/symlink1", "path/to/symlink2"]
move_and_symlink(file_location, directories)
```

```python
import subprocess
from pathlib import Path


def move_file(current, id):
    current = Path(current)
    dest = Path("data_by_id") / id
    try:
        assert dest.exists(), f"Destination path {dest} does not exist."
        assert current.exists(), f"Source file {current} does not exist."
        tar_path = dest / (id + ".tar.xz")
        assert not tar_path.exists(), f"Destination tarball {tar_path} already exists."
        # Build the tar command
        command = [
            "tar",
            "-cJf",
            str(tar_path),
            # "--force",
            "-C",
            str(current.parent),
            str(current.name),
        ]
        # Run the tar process asynchronously
        subprocess.Popen(command, stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)
        print(f"Submitted compression of {current} to {tar_path}.")
    except Exception as e:
        print(f"Failed to move and compress {current} to {dest}: {e}")
```
